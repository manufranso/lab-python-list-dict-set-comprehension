````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
#1

def initialize_inventory(products):
       inventory = {product: int(input(f"Enter a quantity of {product}s avaliable: ")) for product in products}
       
       return inventory   
inventory = initialize_inventory(products)
````

````python
#2

def get_customer_orders():
    customer_orders = int(input("Enter the number of customer orders")) 
    product =[input("Enter the name of a product that a customer wants to order") for product in range(customer_orders) ]
        
    return product

result_orders = get_customer_orders()
````

````python
#3

def calculate_total_price(customer_orders):
    price = [float(input(f"Enter the price of {product}")) for product in customer_orders ]
    total_sum = sum(price)
    return total_sum

result_orders = get_customer_orders()
final_price = calculate_total_price(result_orders)

````

````python
#4

def update_inventory(inventory,customer_orders):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
    inventory =  {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

new_inventory = update_inventory(inventory,result_orders)
````

````python
inventory = initialize_inventory(products)
print("")

customer_orders = get_customer_orders()

print("\nOrders Statistics")
print(f"Total Products Ordered: {len(customer_orders)}")

percentage = (len(set(customer_orders)) / len(products)*100)
print(f"Percentage of Unique Products Ordered : {percentage}")

inventory = update_inventory(inventory,customer_orders)

print("\nUpdated Inventory: ")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")


final_price = calculate_total_price(customer_orders)
print(f"Total Price: {final_price}")
````
